### 8. Visualize how popularity of genres has changed over the years

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use `sum(axis=1)` store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.  

```python
df4=df2
df4=df4.drop(['movie id','movie title','release date','unknown'],axis=1) #dropping unnecessary columns
df4.set_index('release year',inplace=True) #set release year as index
df4
```

```python
grouped_data_2 = df4.groupby('release year') #grouping the baesd on release year
print(grouped_data_2)
```

```python
d=grouped_data_2.sum()  #sum of genres for every year
d
```

```python
d["total"] = d.sum(axis=1)  #creating a new column that contains total genres released every year
d.tail()
```

```python
e=d.div(d['total'], axis= 0) * 100 #percentage of each genre released every year
e.tail() #last 5 cells in a dataframe
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(e, annot=True)
plt.show() #heatmap to analyse popularisation of genres over the years
```

### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
df5=pd.merge(df1,df2,how='right',on='movie id') #merge data and item dataframe
df5
```

```python
df5=df5.drop(['user id','movie id','timestamp','release date','unknown','Action','Adventure','Animation','Childrens','Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western','release year'],axis=1)
df5 # dropping unnecessary columns
```

```python
df5.isnull().sum().sum() #Checking Null values
```

```python
grouped_data_3 = df5.groupby('movie title')  #Grouping the data based on movie title
print(grouped_data_3)
```

```python
f=grouped_data_3.mean() # Average ratings of each movie
f
```

```python
f.reset_index() #resetting index
```

```python
g=grouped_data_3.sum() #count of ratings for each movie
g
```

```python
h=g[g['rating'] <= 100].index #find movies whose ratings less than or equal to hundred
h
```

```python
g.drop(h, inplace = True) #drop movies of ratings less than or equal to 100
g
```

```python
g.reset_index() #resetting index
```

```python
l=pd.merge(g,f,how='left',on='movie title') #merging g and f to get avarage rating and number of rating in single dataframe
l
```

```python
l.isnull().sum().sum() #checking null values
```

```python
l = l.rename(columns={'rating_x': 'rating','rating_y':'Average rating'}) #renaming the columns
l
```

```python
l.sort_values(by='Average rating',ascending= False,inplace=True) #Sort values according to average rating such that movie which has highest rating is on top
l
```

```python
l.iloc[0:25] #selecting top 25 movies based on average ratings where each movie has more than 100 ratings
```
